### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup as BS

url = 'https://www.ettoday.net/news/news-list.htm'
```

```python
resp = requests.get(url)
soup = BS(resp.text)

for d in soup.find(class_ = 'part_list_2').find_all('h3'):
    print(d.find(class_="date").text, d.find_all('em')[-1].text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import json

Ans2 = {}

for d in soup.find(class_ = 'part_list_2').find_all('h3'):
    date = d.find(class_="date").text
    category = d.find(class_ = 'tag').text
    title = d.find_all('a')[-1].text
    Ans2.setdefault(category, [])
    Ans2[category].append({'date':date, "title":title})
print(Ans2)
```

### ③ 哪一個類別的文章最多

```python
iamlist=[]
dic = {}
for x in Ans2:
    num = len(Ans2[x])
    print(x,num)
    
    dic.update({x:num})
iamlist.append(dic)

print (sorted(dic.items(), key=lambda d: d[-1],reverse = True))

```
